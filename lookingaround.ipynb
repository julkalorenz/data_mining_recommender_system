```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies_df = pd.read_csv("data/movies.csv")
ratings_df = pd.read_csv("data/ratings.csv").drop("timestamp", axis=1)
movies_df
ratings_df.loc[ratings_df["movieId"] == 8738]

```

```python

test_ratings_df = ratings_df.sample(int(len(ratings_df)*0.1), random_state=420)
sampled_indices = test_ratings_df.index

test_pairs =  test_ratings_df[["userId", "movieId"]]
actual_ratings = test_ratings_df["rating"]

actual_ratings
```

```python
# Dropping the sampled rows to get the remaining ones
remaining_df = ratings_df.drop(sampled_indices)
remaining_df
```

```python
X = pd.pivot_table(remaining_df, index=["userId"], columns="movieId").applymap(lambda x: not np.isnan(x))
X
```

```python
from sklearn.cluster import KMeans

```

```python
inertias = []
Ks = range(1,57, 2)
for k in Ks:
    kmeans = KMeans(n_clusters=k, n_init="auto", random_state=0xc00ff3).fit(X)
    inertias.append(kmeans.inertia_)
    
sns.lineplot(x=Ks, y=inertias)

```

```python
best_k = 29
kmeans = KMeans(n_clusters=best_k, n_init="auto", random_state=0xc00ff3).fit(X)

users_groups = pd.DataFrame({"cluster": kmeans.labels_}, index=X.index,)
remaining_df["cluster"] = remaining_df["userId"].map(lambda id: int(users_groups.loc[users_groups.index == id]["cluster"]))
```

```python
average_ranking = remaining_df["rating"].mean()
def predict(user_id, movie_id):
    cluster = int(users_groups.iloc[users_groups.index == user_id]["cluster"])
#     same_group = users_groups.loc[users_groups["cluster"] == cluster]
    
#     ratings = remaining_df[(remaining_df["userId"].isin(same_group.index)) & (remaining_df["movieId"] == movie_id)]
#     return ratings["rating"].mean()
        
    rating = remaining_df.loc[(remaining_df["cluster"] == cluster) & (remaining_df["movieId"] == movie_id)]["rating"].mean()
    
    if np.isnan(rating):
        rating = remaining_df.loc[remaining_df["movieId"] == movie_id]["rating"].mean()
        
        if np.isnan(rating):
            rating = average_ranking
        
    return rating
    
y_predicted = np.array([predict(user_id, movie_id) for user_id, movie_id in zip(test_pairs["userId"] ,test_pairs["movieId"])])

y_predicted
    
mean_squared_error(y_predicted, actual_ratings)
```
